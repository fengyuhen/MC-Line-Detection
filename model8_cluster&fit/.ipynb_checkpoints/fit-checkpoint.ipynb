```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import os
import json
from tqdm import tqdm
import random
import warnings

%matplotlib inline
DATA_SHAPE = 224
warnings.simplefilter("ignore", np.RankWarning)
val_labels = np.load('/home/mc16/pre_data/val_label_%s.npy'%DATA_SHAPE)
val_cluster_points = np.load('/home/mc16/result/val_cluster_06-13-01:38:14.npy')
test_cluster_points = np.load('/home/mc16/result/test_cluster_06-13-02:30:18.npy')
```

```python
def cluster_to_img(cluster_points, shape):
    pic = np.zeros((shape, shape), np.uint8)
    num_label = cluster_points.shape[0]
    for i in range(num_label):
        for point in cluster_points[i]:
            pic[point[0], point[1]] = 50 + i * int((255-50) / num_label)
    return pic

test_index = random.randint(0, len(val_cluster_points))
test_label = val_labels[test_index,:,:,1]
test_cluster = val_cluster_points[test_index]
cluster_img = cluster_to_img(test_cluster, DATA_SHAPE)
fig, axs = plt.subplots(1, 2, figsize=(15,15))
axs[0].imshow(test_label, cmap='gray')
axs[1].imshow(cluster_img, cmap='gray')
plt.show()
```

```python
def get_xy(one_cluster):
    x = np.zeros(one_cluster.shape[0], dtype=np.uint8)
    y = np.zeros(one_cluster.shape[0], dtype=np.uint8)
    for i, point in enumerate(one_cluster):
        x[i] = point[0]
        y[i] = point[1]
    return np.clip(x, 0, 223), np.clip(y, 0, 223)

def fit_line(x, y, d):
    zxy = np.polyfit(x, y, d)
    pxy = np.poly1d(zxy)
    zyx = np.polyfit(y, x, d)
    pyx = np.poly1d(zyx)
    return pxy, pyx

def extend_line(x, y, pxy, pyx):
    x_ = x
    if(pxy[1] < 0): # left line
        if(pyx(0) < 224):
            print("pyx(0) < 224", pyx(0))
            extend = np.array(range(max(x), int(pyx(0))))
            x_ = np.append(x ,extend)
            print(max(x_))
        else:
            print("pyx(0) > 224", pyx(0))
            extend = np.array(range(max(x), 224))
            x_ = np.append(x ,extend)
            print(max(x_))
            
    else:              # right line
        if(pyx(224) < 224):
            print("pyx(224) < 224", pyx(224))
            extend = np.array(range(max(x), int(pyx(224))))
            x_ = np.append(x, extend)
            print(max(x_))
        else:
            print("pyx(224) > 224", pyx(224))
            extend = np.array(range(max(x), 224))
            x_ = np.append(x, extend)
            print(max(x_))
    return x_
        
# get one line from one cluster
def get_line(one_cluster, d):
    line = np.zeros((len(one_cluster), 2), dtype = np.uint8)
    x, y = get_xy(one_cluster)
    if(len(x)==0 | len(y)==0):
        return line
    pxy, pyx = fit_line(x, y, d)
    new_x = extend_line(x, y, pxy, pyx)
    new_y = pxy(new_x)
    line = np.zeros((len(new_y), 2), dtype = np.uint8)
    for i in range(len(new_x)): # get a new line
        line[i, 0] = new_x[i].astype(np.uint8)
        line[i, 1] = new_y[i].astype(np.uint8)
    return line

# get lines from clusters for a image
def get_lines(clusters, d):
    lines = []
    for cluster in clusters:
        if(len(cluster) <= 20): # delete noise
            continue
        line = get_line(cluster, d)
#         print(line)
        lines.append(line)
    return lines

lines = get_lines(test_cluster, 1)
```

```python
def fit_points(points):
    line_info = []
    fig = plt.figure()
    fig.gca().invert_yaxis()
    ax1 = fig.add_subplot(111)
    ax1.set_ylim(0, 224)
    ax1.set_xlim(0, 224)
    line_num = points.__len__()
    temp = []
    for i in range(points.__len__()):
        x = points[i][:, 0]; y = points[i][:, 1]
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        y_val = p(x)
        temp_x = range(112, 224)
        temp_y = p(temp_x)
        temp_x_y = []
        for j in range(temp_x.__len__()):
            temp_x_y.append([temp_x[j], int(temp_y[j])])
        temp.append([i+1, temp_x_y])
        plot1 = plt.plot(y, x, '*')
        plot2 = plt.plot(temp_y, temp_x)
    plt.show()
    line_info = [line_num, temp]
    return line_info

extend = fit_points(test_cluster)
print(extend[0])
```

```python
def lines_to_img(lines, shape):
    pic = np.zeros((shape, shape), np.uint8)
    for line in lines:
        for point in line:
            pic[point[0], point[1]] = 255
    return pic

# test_index = random.randint(0, len(val_cluster_points))
test_index = 1934
test_label = val_labels[test_index,:,:,1]
test_cluster = val_cluster_points[test_index]
cluster_img = cluster_to_img(test_cluster, DATA_SHAPE)
test_lines = get_lines(test_cluster, 1)
lines_img = lines_to_img(test_lines, DATA_SHAPE)
fig, axs = plt.subplots(1, 3, figsize=(20,20))
axs[0].imshow(test_label, cmap='gray')
axs[1].imshow(cluster_img, cmap='gray')
axs[2].imshow(lines_img, cmap='gray')
print(test_index, len(test_lines))
plt.show()
```

```python
# get lines for all images in a dataset
def get_lines_set(clusters_list, d):
    lines_list = []
    for clusters in tqdm(clusters_list):
        lines = get_lines(clusters, d)
        lines_list.append(lines)
    return lines_list
    
val_lines = get_lines_set(val_cluster_points, 2)
# test_lines = get_lines_set(test_cluster_points, 2)
```

# Save json

```python
import json
def np2json(arlines):
    json_lines = []
    for arline in arlines:
        json_line = []
        for point in arline:
            json_line.append({"x":int(point[1]), "y":int(point[0])})
        json_lines.append(json_line)
    return {"Lanes": json_lines}

def save_jsons(lines_list, list_path, json_path):
    test_md5_paths = open(list_path, 'r').readlines()
    for i, lines in tqdm(enumerate(lines_list)):
        md5 = test_md5_paths[i].split('.')[0]
        lines_json = np2json(lines)
        lines_str = json.dumps(lines_json)
        file = open(json_path + '%s.json'%md5,'w')
        file.write(lines_str)
        file.close()
        
test_list_path = '/home/mc16/test0_list.txt'
test_json_path = '/home/mc16/result_json/results/'
val_list_path = '/data/mc_data/MLDC/data/val/list.txt'
val_json_path = '/home/mc16/result_json/val_results/'
# save_jsons(test_lines, test_list_path, test_json_path)
save_jsons(val_lines, val_list_path, val_json_path)
```

```python
from keras.utils import to_categorical
test_images = np.load('/home/mc16/pre_data/test_image_224.npy')
```

```python
def read_paths(list_path, result_path):
    test_md5_paths = open(list_path,'r').readlines()
    labelpaths = []
    for path in tqdm(test_md5_paths):
        md5 = path.split('.')[0]
        labelpaths.append(result_path + '%s.json'%md5)
    return labelpaths

def get_label_json(label_paths):
    labels_json = []
    for labelpath in tqdm(label_paths):
        with open(labelpath, 'r') as f:
            data = json.load(f)
            labels_json.append(data)
    return labels_json

def get_origin_label(labels_json, shape):
    labels = np.zeros((len(labels_json), shape, shape), np.uint8)
    for i in tqdm(xrange(len(labels_json))):
        for j,line in enumerate(labels_json[i].values()[0]):
            for k, point in enumerate(line):
                x, y = int(float(point['x']+.5)), int(float(point['y']+0.5))
                if(x >= shape): x = shape-1
                if(y >= shape): y = shape-1
                labels[i][y][x] = 1
                if x < shape-1: labels[i][y][x+1] = 1
                if x > 0      : labels[i][y][x-1] = 1
                if y < shape-1: labels[i][y+1][x] = 1
                if y > 0      : labels[i][y-1][x] = 1          
    return labels

def get_mask_label(labels):
    mask_labels = np.zeros((labels.shape[0], labels.shape[1], labels.shape[2], 2), np.uint8)
    for i in tqdm(xrange(len(labels))):
        mask_labels[i] = to_categorical(labels[i], num_classes=2)
    return mask_labels

def get_label(label_paths, shape):
    label_json = get_label_json(label_paths)
    origin_label = get_origin_label(label_json, shape)
    mask_label = get_mask_label(origin_label)
    return mask_label

test_list_path = '/home/mc16/test0_list.txt'
test_result_path = '/home/mc16/result/06111/'
test_result_md5s = read_paths(test_list_path, test_result_path)
test_results = get_label(test_result_md5s, DATA_SHAPE)
```
