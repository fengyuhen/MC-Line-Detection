# Line classifer
- 1：一条直线（一次函数拟合并延长）
- 2：一条曲线（按X轴取像素中心拟合）
- 3：Y型线&顶部粘住的两条线（分两条按X轴取像素中心拟合
- 4：无效线（小圆块，大散块，多条线粘一起等）（不拟合）

```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import random
TIMEFORMAT = "%m-%d-%H:%M:%S"
val_cluster = np.load('/home/mc16/result/val_cluster_06-18-05:50:42.npy')
```

```python
classify_line_list = [[] for i in range(4)]

def line2img(line):
    img = np.zeros((224, 224), np.uint8)
    for point in line:
        img[point[0], point[1]] = 1
    return img

def judge(lines_list):    
    while True:
        index = random.randint(0, len(lines_list)-1)
        lines = lines_list[index]
        for line in lines[0]:
            img = line2img(line)
            plt.imshow(img, cmap='gray')
            plt.show()
            line_class = input('line class: ')
            classify_line_list[line_class-1].append(line)
            print(len(classify_line_list[0])+len(classify_line_list[1])+len(classify_line_list[2])+len(classify_line_list[3]))
            
            
judge(val_cluster)
```

```python
print(len(classify_line_list[0]),len(classify_line_list[1]),len(classify_line_list[2]),len(classify_line_list[3]))
print(len(classify_line_list[0])+len(classify_line_list[1])+len(classify_line_list[2])+len(classify_line_list[3]))
time = datetime.datetime.now().strftime(TIMEFORMAT)
np.save('line_class_%s'%time, classify_line_list)
```

# line show and augmentation

```python
from imgaug import augmenters as iaa
import cv2
from sklearn.utils import shuffle
lines = np.load('/home/mc16/model9_lineclass/line_class_06-21-00:04:18.npy')
DATA_SHAPE = 32

def line2img(line, shape):
    x_list = line[:, 0]
    y_list = line[:, 1]
    x_range = x_list.max() - x_list.min() + 1
    y_range = y_list.max() - y_list.min() + 1
    x_list = x_list - x_list.min()
    y_list = y_list - y_list.min()
    pic = np.zeros((x_range, y_range), np.uint8)
    for x, y in zip(x_list, y_list):
        pic[x, y] = 255
    img = cv2.resize(pic ,(shape, shape))
    return img

def lines2imgs(lines, shape):
    imgs = []
    labels = []
    for i in range(len(lines)):
        for j in range(len(lines[i])):
            img = line2img(lines[i][j], shape)
            imgs.append(img)
            labels.append(i)
    return imgs, labels

imgs, labels = lines2imgs(lines, DATA_SHAPE)
```

```python
# print(len(imgs), len(labels))
# print(len(lines[0]), len(lines[1]), len(lines[2]), len(lines[3]))
fliplr = iaa.Fliplr(1.0)
flipud = iaa.Flipud(1.0)

def show(imgs, labels):
    show_index = random.randint(0, len(imgs))
    origin_img = imgs[show_index]
    show_label = labels[show_index]
    fliplr_img = fliplr.augment_image(origin_img)
    flipud_img = flipud.augment_image(origin_img)
    print(show_label + 1)

    fig, axs = plt.subplots(1, 3, figsize=(15,5))
    axs[0].set_title('origin', fontsize=18), axs[0].imshow(origin_img, cmap='gray')
    axs[1].set_title('fliplr', fontsize=18), axs[1].imshow(fliplr_img, cmap='gray')
    axs[2].set_title('flipud', fontsize=18), axs[2].imshow(flipud_img,cmap='gray')
    plt.show()
    
show(imgs, labels)
```

```python
def lines_2class_imgs(lines, label, shape):
    imgs = []
    labels = []
    for line in lines:
        img = line2img(line, shape)
        imgs.append(img)
        labels.append(label)
    return imgs, labels

def get_aug_imgs(imgs, labels, p):
    aug_imgs = []
    aug_labels = []
    for img, label in zip(imgs, labels):
        if random.random() < p:
            fliplr_img = fliplr.augment_image(img)
            aug_imgs.append(fliplr_img)
            aug_labels.append(label)
        if random.random() < p:
            flipud_img = flipud.augment_image(img)
            aug_imgs.append(flipud_img)
            aug_labels.append(label)
    aug_imgs, aug_labels = shuffle(aug_imgs + imgs, aug_labels + labels)        
    return aug_imgs, aug_labels

line1_imgs, line1_labels = lines_2class_imgs(lines[0], 0, DATA_SHAPE)
print(len(lines[0]), len(line1_imgs), len(line1_labels))
line2_imgs, line2_labels = lines_2class_imgs(lines[1], 1, DATA_SHAPE)
aug2_imgs, aug2_labels = get_aug_imgs(line2_imgs, line2_labels, 1.)
print(len(lines[1]), len(aug2_imgs), len(aug2_labels))
line3_imgs, line3_labels = lines_2class_imgs(lines[2], 2, DATA_SHAPE)
aug3_imgs, aug3_labels = get_aug_imgs(line3_imgs, line3_labels, 1.)
print(len(lines[2]), len(aug3_imgs), len(aug3_labels))
line4_imgs, line4_labels = lines_2class_imgs(lines[3], 3, DATA_SHAPE)
aug4_imgs, aug4_labels = get_aug_imgs(line4_imgs, line4_labels, 1.)
print(len(lines[3]), len(aug4_imgs), len(aug4_labels))
aug_imgs = line1_imgs + aug2_imgs + aug3_imgs + aug4_imgs
aug_labels = line1_labels + aug2_labels + aug3_labels + aug4_labels
print(len(aug_imgs), len(aug_labels))
aug_imgs, aug_labels = shuffle(aug_imgs, aug_labels)
```

```python
show(aug_imgs, aug_labels)
time = datetime.datetime.now().strftime(TIMEFORMAT)
np.save('aug_class_images_%s'%time, aug_imgs)
np.save('aug_class_labels_%s'%time, aug_labels)
```
